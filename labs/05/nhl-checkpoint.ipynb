```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


   
```

```python
table2=[]
number=1
while True:
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={number}"  #url hochzählen für die anderen Seiten 
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    table = soup.find_all('th')                                      #die Überschriften finden 
    table_collums = [collums.text.strip() for collums in table]      #die Überschriften als Text und ohne Leehrzeichen


    column_data = soup.find_all('tr')                                #alle Zeilen finden 
    if len(column_data) <= 1:                                        #suche abbrechen, wenn die Tabelle keine Inhalte mehr hat außer der Überschriften
        break
    for row in column_data[1:]:                                     #die Zeilen durchgehen 
        row = row.find_all('td')                                    #in den Zielen die Inhalte durchgehen
        ind_row_data = [data.text.strip() for data in row]          #den Zeileninhalt als Text und ohne Leehrzeichen
        table2.append(ind_row_data)                                 


    df = pd.DataFrame(columns=table_collums,data = table2)         #mit pandas eine Tabelle erstellen 
    number+=1

df.to_csv(r'C:\Users\lenem\Documents\Dokumente\TH\3_Semester\Data_Modelling\scraping.csv', index = False )     #datei als csv speichern
```

```python
years = ['1990','2000','2010']                                                          #Jahre definieren
output = df[df['Year'].isin (years)]                                                    #Jahre Filtern
first_team = output.sort_values('Wins', ascending=False).groupby('Year').head(1)        #nach Gewinnen sortiern und 1. Platz ausgeben lassen
print(first_team[['Wins','Team Name','Year']])
```
